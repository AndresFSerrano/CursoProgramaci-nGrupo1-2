```python
import csv
import json
def clima():
    
    archivo=open("data.csv")
    x=csv.reader(archivo)
    
    #Creamos las listas vacías
    id=[]
    location=[]
    temperature=[]
    pressure=[]
    
    next(x,None)
    for linea in x:
      id.append(linea[0])
      location.append(linea[1])
      temperature.append(int(linea[2]))
      pressure.append(int(linea[3]))
    
    temperatura1=[]
    presion1=[]
    temperatura2=[]
    presion2=[]
    temperatura3=[]
    presion3=[]
    temperatura4=[]
    presion4=[]
    temperatura5=[]
    presion5=[]
    
    for i,k in enumerate(location):
          #i es la posición
          #k es el valor de la locación
          if k=='1':
            temperatura1.append(temperature[i])
            presion1.append(pressure[i])
          if k=='2':
            temperatura2.append(temperature[i])
            presion2.append(pressure[i])
          if k=='3':
            temperatura3.append(temperature[i])
            presion3.append(pressure[i])
          if k=='4':
            temperatura4.append(temperature[i])
            presion4.append(pressure[i])
          if k=='5':
            temperatura5.append(temperature[i])
            presion5.append(pressure[i])
            
          
    avgt1=round(sum(temperatura1)/len(temperatura1),1)
    avgp1=round(sum(presion1)/len(presion1),1)
    
    avgt2=round(sum(temperatura2)/len(temperatura2),1)
    avgp2=round(sum(presion2)/len(presion2),1)
    
    avgt3=round(sum(temperatura3)/len(temperatura3),1)
    avgp3=round(sum(presion3)/len(presion3),1)
    
    avgt4=round(sum(temperatura4)/len(temperatura4),1)
    avgp4=round(sum(presion4)/len(presion4),1)
    
    avgt5=round(sum(temperatura5)/len(temperatura5),1)
    avgp5=round(sum(presion5)/len(presion5),1)
    
    archivo.close()
    data={}
    #Alimentamos el diccionario
    data["1"]=[avgt1,avgp1]
    data["2"]=[avgt2,avgp2]
    data["3"]=[avgt3,avgp3]
    data["4"]=[avgt4,avgp4]
    data["5"]=[avgt5,avgp5]
    
    with open('jason.json','w') as file:
      json.dump(data,file)
    
    with open('jason.json','r') as file:
      v=json.load(file)
    
    datos=json.dumps(v)
    
    with open("data_nuevo.csv", "w") as file:   #Lo mismo que file=open("ruta_archivo", "w")
        file.write("id,")
        file.write("location,")
        file.write("temperature,")
        file.write("pressure,")
        file.write("above_avg_temp,")
        file.write("above_avg_pres\n")
        n=len(id)
        
        above_avg_temp=[]
        above_avg_pres=[]
        for i in range(n):
          if location[i]=="1":
            if temperature[i]>avgt1:
                above_avg_temp.append('SI')
            elif temperature[i]<avgt1:
                above_avg_temp.append('NO')
            elif temperature[i]==avgt1:
                above_avg_temp.append('IGUAL')
            if pressure[i]>avgp1:
                above_avg_pres.append('SI')
            elif pressure[i]<avgp1:
                above_avg_pres.append('NO')
            elif pressure[i]==avgp1:
                above_avg_pres.append('IGUAL')
          if location[i]=="2":
            if temperature[i]>avgt2:
                above_avg_temp.append('SI')
            elif temperature[i]<avgt2:
                above_avg_temp.append('NO')
            elif temperature[i]==avgt2:
                above_avg_temp.append('IGUAL')
            if pressure[i]>avgp2:
                above_avg_pres.append('SI')
            elif pressure[i]<avgp2:
                above_avg_pres.append('NO')
            elif pressure[i]==avgp2:
                above_avg_pres.append('IGUAL')
          if location[i]=="3":
            if temperature[i]>avgt3:
                above_avg_temp.append('SI')
            elif temperature[i]<avgt3:
                above_avg_temp.append('NO')

            if pressure[i]>avgp3:
                above_avg_pres.append('SI')
            elif pressure[i]<avgp3:
                above_avg_pres.append('NO')
            elif pressure[i]==avgp3:
                above_avg_pres.append('IGUAL')
          if location[i]=="4":
            if temperature[i]>avgt4:
                above_avg_temp.append('SI')
            elif temperature[i]<avgt4:
                above_avg_temp.append('NO')
            elif temperature[i]==avgt4:
                above_avg_temp.append('IGUAL')
            if pressure[i]>avgp4:
                above_avg_pres.append('SI')
            elif pressure[i]<avgp4:
                above_avg_pres.append('NO')

          if location[i]=="5":
            if temperature[i]>avgt5:
                above_avg_temp.append('SI')
            elif temperature[i]<avgt5:
                above_avg_temp.append('NO')
            elif temperature[i]==avgt5:
                above_avg_temp.append('IGUAL')
            if pressure[i]>avgp5:
                above_avg_pres.append('SI')
            elif pressure[i]<avgp5:
                above_avg_pres.append('NO')

        n=len(id)
        for i in range(n):
              file.write("%s," % id[i])  
              file.write("%s," % location[i])
              file.write("%s," % temperature[i])
              file.write("%s," % pressure[i])
              file.write("%s," % above_avg_temp[i])
              file.write("%s\n" % above_avg_pres[i])
        
        file.close()  #Se cierra la variable temporal file creada  

    return datos
```

```python
clima()
```
